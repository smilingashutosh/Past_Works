#Start the Visualization

```python
# generate the data points

import numpy as np
X = np.linspace(-np.pi,np.pi,256)
s = np.sin(X)
c = np.cos(X)

import matplotlib.pyplot as plt
# print(X)
# print(s)
plt.plot(X,s)
plt.plot(X,c)
```
